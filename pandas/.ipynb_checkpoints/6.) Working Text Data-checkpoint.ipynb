# Filtering using String Methods

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Departemrnt"] = chicago["Department"].astype("category")
chicago.tail(4)
```

```python
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
chicago[chicago["Position Title"].str.lower().str.startswith("water")]
```

```python
chicago[chicago["Position Title"].str.lower().str.endswith("ist")]
```

# Split String by Charecters with .str.split() 

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Departemrnt"] = chicago["Department"].astype("category")
chicago.head(4)
```

```python
chicago["Name"].str.split(",")
```

```python
chicago["Name"].str.split(",").str.get(0).str.title().value_counts()
```

```python
chicago["Name"].str.split(",").str.get(1).str.title().value_counts().head()
```

```python
chicago["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).value_counts().head()
```

# The Expand and n Parameters of the str.split()

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Departemrnt"] = chicago["Department"].astype("category")
chicago.head(4)
```

```python
chicago["Name"].str.split(",",expand = True)
```

```python
chicago[["First Name","Last Name"]] = chicago["Name"].str.split(",",expand = True)
```

```python
chicago.head()
```

```python
chicago[["Position part 1","Position part 2"]] = chicago["Position Title"].str.split(" ",expand = True , n=1)
```

```python
chicago.head()
```
