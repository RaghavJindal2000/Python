# Review of the .apply() Method on Single Columns

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(2)
```

```python
def Convert_and_add(number):
    return str(number) + "MILLIONS!"

bond["Box Office"] = bond["Box Office"].apply(Convert_and_add)
```

```python
bond.head(3)
```

```python
columns = ["Budget","Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(Convert_and_add)
```

```python
bond.head(4)
```

# The .apply() Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(2)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor=="Pierce Bronson":
        return "The BEST"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I dont Know"

bond.apply(good_movie,axis="columns")
```
